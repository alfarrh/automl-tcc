```python
import pandas as pd
from scipy.io import arff

from _libs.paths import PATHS
```

```python
data, meta = arff.loadarff(PATHS["KIDNEY_DISEASE"])
```

```python
df = pd.DataFrame(data)
df = df.map(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
df.to_csv(PATHS["KIDNEY_DISEASE"].replace(".arff", ".csv"), index=False)
```
